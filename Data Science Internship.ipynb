```python
df_2019 = df[df['season']==2019]
```

```python
df_2019.columns
```

```python
ans=pd.DataFrame(columns=['batsman','total_runs','num_of_dotballs','runs_scored_by_taking_1'
                          ,'runs_scored_by_taking_2','runs_scored_by_taking_3','runs_scored_by_taking_4','runs_scored_by_taking_6'])
```

```python
for (key,data) in df_2019.groupby(['batsman']):
    d={}
    d['batsman']=key
    d['total_runs']=data['batsman_runs'].sum()
    d['num_of_dotballs']=(data['batsman_runs']==0).sum()
    d['runs_scored_by_taking_1']=(data['batsman_runs']==1).sum()*1
    d['runs_scored_by_taking_2']=(data['batsman_runs']==2).sum()*2
    d['runs_scored_by_taking_3']=(data['batsman_runs']==3).sum()*3
    d['runs_scored_by_taking_4']=(data['batsman_runs']==4).sum()*4
    d['runs_scored_by_taking_6']=(data['batsman_runs']==6).sum()*6
    ans=pd.concat([ans,pd.DataFrame(d,index=[0])])
```

```python
ans['total_runs_by_power_hiting'] = ans['runs_scored_by_taking_4'] + ans['runs_scored_by_taking_6']
```

```python
ans.loc[:,['batsman','total_runs_by_power_hiting']].sort_values(by='total_runs_by_power_hiting',ascending=False)
```

```python
ans=pd.DataFrame(columns=['batsman','number of balls faced','number of 4\'s and 6\'s'])
```

```python
hard_ability = df.loc[:,['batsman','batsman_runs']]
```

```python
for (k,d) in hard_ability.groupby('batsman'):
    dic={}
    dic['batsman']=k
    dic['number of balls faced']=d.shape[0]
    dic['number of 4\'s and 6\'s']=d[d['batsman_runs'].isin([4,6])].shape[0]
    ans = pd.concat([ans,pd.DataFrame(dic,index=[0])],ignore_index=True)
```

```python
ans['hard'] = ans['number of 4\'s and 6\'s']/ans['number of balls faced']
```

```python
ans[ans['batsman'].isin(['V Kohli','AB de Villiers','MS Dhoni','CH Gayle'])].sort_values(by='hard',ascending=False)
```
